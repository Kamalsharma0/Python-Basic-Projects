# Student Result Predicton (Pass/Fail)

Step 1 : Taake training data

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Kamalsharma0/Python-/refs/heads/main/Data%20Sets/iris.csv"
```

```python
# read the csv file using
df = pd.read_csv(url)
print(df)
```

Step 2 : Vertically devide the data into Input(X) and Output(Y)

```python
df1 = df.values
x = df1[:,0:2]
y = df1[:,2]
```

```python
y[:5]
```

Step 3 : Choose appropriate Machine Learning Algorithms/model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Step 4 : Fit the data into the selected model

```python
model.fit(x,y)
```

Step 5 : Predict the results with user input

```python
# user inputs
self_study = float(input("Enter the number of hours self studied? "))
tutorial_study = float(input("Enter the number of hours tutorials you watched? "))

# print output
print(f"Predicted Output (Pass=1, Fail=0): {model.predict([[self_study, tutorial_study]])}")
```
